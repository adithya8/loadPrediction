```python
import pandas as pd
import numpy as np
from keras.models import Sequential
from keras.layers import Dense
from sklearn.model_selection import train_test_split
import matplotlib as plt

path = "J:\\SF\\Thermal Load Prediction\\Datas\\Low Carbon London 168 homes\\"
read = pd.read_csv(path+"Power-Networks-LCL-June2015(withAcornGps)v2_1.csv")
read = read[read.columns[3:4]]

def coeff_determination(y_true, y_pred):
    SS_res =  K.sum(K.square( y_true-y_pred )) 
    SS_tot = K.sum(K.square( y_true - K.mean(y_true) ) ) 
    return ( 1 - SS_res/(SS_tot + K.epsilon()) )

ndf = pd.concat([read.shift(4), read.shift(2), read], axis=1)
ndf.columns = ('t-2', 't-1', 't')
ndf = ndf.iloc[4:]
sh = int(ndf.shape[0]*.20)

x = ndf[ndf.columns[0:2]]
y = ndf[ndf.columns[-1:]]

for i in range(x.shape[0]):
    if(x.values[i][0]=='Null'):
        x.values[i][0]='0'
    elif(x.values[i][1]=='Null'):
        x.values[i][1]='0'
    elif(y.values[i][0]=='Null'):
        y.values[i][0]='0'

x = np.array(x, dtype=float)
y = np.array(y, dtype=float)

act = ['linear','sigmoid','tanh','softsign']
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=1)
for i in range(4):
    print(act[i])
    model = Sequential()
    model.add(Dense(5,input_dim=2,kernel_initializer='normal', activation=act[i]))
    #Output Layer.
    model.add(Dense(1,kernel_initializer='normal'))
    #Compile model.
    model.compile(loss='mean_squared_error',optimizer='adam',metrics=[coeff_determination])
    #Fit the model.
    his = model.fit(X_train,y_train,epochs=3,batch_size=200,verbose=2)
    print(his)
    #Final Evaluation of the model.
    scores = model.evaluate(X_test,y_test,verbose=0)
    print ("Baseline Error : %.5f" % scores[i])
    model.summary()
```
